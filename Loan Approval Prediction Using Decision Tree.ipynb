```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

data = {
    "Gender": ["Male", "Female", "Male", "Female", "Male", "Female"],
    "Married": ["Yes", "No", "Yes", "No", "Yes", "Yes"],
    "Income": [500000, 300000, 650000, 320000, 480000, 290000],
    "Credit_Score": [750, 580, 720, 600, 680, 590],
    "Loan_Status": ["Y", "N", "Y", "N", "Y", "N"]
}

df = pd.DataFrame(data)

le_gender = LabelEncoder()
le_married = LabelEncoder()
le_loan = LabelEncoder()

df["Gender"] = le_gender.fit_transform(df["Gender"])
df["Married"] = le_married.fit_transform(df["Married"])
df["Loan_Status"] = le_loan.fit_transform(df["Loan_Status"])

x = df[["Gender", "Married", "Income", "Credit_Score"]]
y = df["Loan_Status"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

model = DecisionTreeClassifier()
model.fit(x_train, y_train)

g = input("Enter Gender(Male/Female): ")
m = input("Married?(Yes/No): ")
inc = float(input("Enter Annual Income: "))
cs = int(input("Enter Credit Score(300-900): "))

g_enc = le_gender.transform([g])[0]
m_enc = le_married.transform([m])[0]

sample = [[g_enc, m_enc, inc, cs]]

pred = model.predict(sample)
result = le_loan.inverse_transform(pred)

print("\nLoan Status Prediction:", result[0])
```
