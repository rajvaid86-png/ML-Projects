```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB

data = {
    'message': [
        "Congratulations! You won a lottery of $1000",
        "Win cash prizes now", 
        "Call this number to claim your reward",
        "Hey, are we meeting today?",
        "Your order has been shipped",
        "Let's go for lunch",
        "Free entry in mega contest",
        "Can you send me the report?"
    ],

    'label': [
        "spam", "spam", "spam",
        "not_spam", "not_spam", "not_spam",
        "spam", "not_spam"
    ]
}

df = pd.DataFrame(data)

vectorizer = CountVectorizer()
x = vectorizer.fit_transform(df['message'])
y = df['label']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

model = MultinomialNB()
model.fit(x_train, y_train)

usr_msg = input("Enter your Message: ")
usr_msg_vec = vectorizer.transform([usr_msg])

prediction = model.predict(usr_msg_vec)[0]
print("\nYour message:", usr_msg)
print("Prediction:", prediction)
```
